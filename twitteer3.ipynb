```python
import snscrape.modules.twitter as sntwitter
import itertools
import re
import numpy as np
import pandas as pd
from textblob import TextBlob
import nltk
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = '"#IPL2019 near:India" since:2019-01-01 until:2021-01-01 lang:en'
maxTweets = 3000
tweetList = []
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(data).get_items()):
    if i > maxTweets:
         break 
    tweetList.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.retweetCount, tweet.user.followersCount])
```

```python
India = pd.DataFrame(tweetList, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'LikeCount', 'Retweet', 'Followers'])
India['location'] = 'India'                                         
India
```

```python
data = '"#IPL2019 near:UK" since:2019-01-01 until:2021-01-01 lang:en'
maxTweets = 3000
tweetList = []
for i,tweet in enumerate(sntwitter.TwitterSearchScraper(data).get_items()):
    if i > maxTweets:
         break 
    tweetList.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.likeCount, tweet.retweetCount, tweet.user.followersCount])
```

```python
uk = pd.DataFrame(tweetList, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'LikeCount', 'Retweet', 'Followers'])
uk['location'] = 'UK'
uk
```

```python
#concatinate both dataframes
df = pd.concat([India, uk], axis=0)
df.reset_index(drop=True, inplace=True)
df
```

## Data Preprocessing


```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub("@[\w]*", "", text)
    text = re.sub("http\S+", "", text)
    text = re.sub("[^a-zA-Z#]", " ", text)
    text = re.sub("rt", "", text)
    text = re.sub("\s+", " ", text)

    return text
```

```python
df['Clean_Text'] = df['Text'].apply(preprocess_text)
df
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def getPolarity(text):
    return TextBlob(text).sentiment.polarity


df['Subjectivity'] = df['Clean_Text'].apply(getSubjectivity)
df['Polarity'] = df['Clean_Text'].apply(getPolarity)
df
```

```python
def getAnalysis(score):
    if score<0:
        return 'Negative'
    elif score == 0:
        return 'Nuetral'
    else:
        return 'Positive'
    
df['Analysis'] = df['Polarity'].apply(getAnalysis)
df
```

```python
df = df.drop(columns=[ 'Datetime', 'Tweet Id', 'Subjectivity', 'Polarity', 'LikeCount', 'Retweet'])
df
```

```python
# Individualism
def Individualness(tweets):
#     mt=0
    mentions = []
    for tweet in tweets:
        it=tweet.split()
        mt=it.count("i")+it.count("me")+it.count("my")+it.count("own")+it.count("able")
        mentions.append(mt)
    #print(mentions)
    return mentions
```

```python
df['individualism'] = Individualness(df['Clean_Text'])
```

```python
#collectivism
def collectiveness(tweets):
#     mt=0
    mentions = []
    for tweet in tweets:
        it=tweet.split()
        mt=it.count("we")+it.count("us")+it.count("our")+it.count("ourselves")+it.count("lets")
        mentions.append(mt)
    #print(mentions)
    return mentions
```

```python
df['collectivism'] = collectiveness(df['Clean_Text'])
df
```

```python
#powerdistance
def tags(tweets):
#     mt=0
    mentions = []
    for tweet in tweets:
        it=tweet.split()
        mt=it.count("@")
        mentions.append(mt)
    #print(mentions)
    return mentions
```

```python
df['Power_distance'] = tags(df['Text'])
df
```

```python
# extract the hashtag
def hashtag_extract(tweets):
    hashtags = []
    # loop words in the tweet
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    return hashtags 
```

```python
ht = hashtag_extract(df['Clean_Text'])
```

```python
ht = sum(ht, [])
```

```python
freq = nltk.FreqDist(ht)
d = pd.DataFrame({'Hashtag': list(freq.keys()),
                 'Count': list(freq.values())})
d.head()
```

```python
# select top 10 hashtags
d = d.nlargest(columns='Count', n=10)
plt.figure(figsize=(15,9))
sns.barplot(data=d, x='Hashtag', y='Count')
plt.show()
```

```python
# individual words considered as tokens
tokenized_tweet = df['Clean_Text'].apply(lambda x: x.split())
tokenized_tweet.head()
```

```python
# stem the words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

tokenized_tweet = tokenized_tweet.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
tokenized_tweet.head()
```

```python
# combine words into single sentence
for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = " ".join(tokenized_tweet[i])
    
df['Clean_Texts'] = tokenized_tweet
df.head()
```
